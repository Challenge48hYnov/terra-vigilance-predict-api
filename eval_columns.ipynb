```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
```

```python
# Chargement des données
df = pd.read_parquet("./datasets/inondation/datasets_inondation.parquet")
y = df["catastrophe"]
X = df.drop(["catastrophe","date"], axis=1)
# Entraînement
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# Importance des features
importances = pd.Series(model.feature_importances_, index=X.columns)
print(importances.sort_values(ascending=False))
```

```python
# Tracer les importances des features sous forme de bar chart
importances.sort_values(ascending=False).plot(kind='bar', figsize=(8, 5))
plt.title("Importances des features")
plt.ylabel("Importance")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
